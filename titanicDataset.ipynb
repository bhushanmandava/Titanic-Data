## Adding missing values in age

```python
guess_age=np.zeros((2,3))
guess_age
```

```python
#finding the the age based on the catogrys sex and classes so we have total 6 comination to play with
#now get the mean of the each vals

for i in range(0,2):
  for j in range(0,3):
    guess_age_val = dataset[(dataset['Sex'] == i) & (dataset['Pclass'] == j+1)]['Age'].dropna()
    potential_age=guess_age_val.mean()
    guess_age[i,j]=potential_age

for i in range(0,2):
  for j in range(0,3):
    dataset.loc[(dataset.Age.isnull()) & (dataset.Sex == i) & (dataset.Pclass == j+1), 'Age'] = guess_age[i,j]
```

```python
dataset.info()
```

## Adding a band range for the Age data

```python
# now i want age band just to make a classification of which age groups have heighest survival rate
#age band classification
#0-10 = 1, 10-20=2,20-30=3,30-40=4,40-50 = 5, 50-60 = 6 ,60-70=7,70- =8
def making_age_band(age) :
  if(age>=0 and age<10):
    return 1
  elif(age>=10 and age<20):
    return 2
  elif(age>=20 and age<30):
    return 3
  elif(age>=30 and age<40):
    return 4
  elif(age>=40 and age<50):
    return 5
  elif(age>=50 and age<60):
    return 6
  elif(age>=60 and age<70):
    return 7
  else:
    return 8
dataset['AgeBand']=dataset['Age'].apply( making_age_band)
```
